```python
#importing libraries 

import numpy as np
import pandas as pd
from os import path
from PIL import Image
from nltk.corpus import stopwords
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#loading the dataset 

df = pd.read_csv('Fortune 500 Mission Statements.csv')
```

```python
df.head()
```

```python
df['Mission'].head()
```

```python
#dropping the missings 

df.dropna(axis=0,how='any') 
```

```python
text = "".join(str(statement) for statement in df.Mission)
```

```python
wordcloud = WordCloud(width=800, height=400).generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud.to_file("missions.png")
```
